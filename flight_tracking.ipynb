```python
import numpy as np
import pandas as pd
import datetime
from datetime import timedelta

def genRefTime(arg_dtStartDate, arg_dtEndDate):
    int_offset  = np.random.randint((arg_dtEndDate - arg_dtStartDate).days)
    dt_refDate = arg_dtStartDate + timedelta(days=int_offset)
    return(dt_refDate + timedelta(seconds=np.random.randint(80000)))
    
def genOneVhcTimeLst(arg_strStartDate, arg_strEndDate, arg_intNum=10):
    lst_dtTime = []
    
    int_offset = np.random.randint(10, 30) 
    dt_refTime = genRefTime(pd.to_datetime(arg_strStartDate),
                            pd.to_datetime(arg_strEndDate))
    n = 0
    while n < arg_intNum:
        dt_oneTime = dt_refTime + timedelta(seconds = (n*int_offset))
        lst_dtTime.append(dt_oneTime)
        n = n +1
    return(lst_dtTime)

def genVehicleDatas(arg_lstKind,
                    arg_lstSize=[10, 50],
                    arg_lstDateRange=["2019-12-01", "2019-12-30"],
                    arg_lstPos=[37.45, 127.01]):
    
    # empty Dataframe 생성
    # "time","vehicleNo", "vehicleName","latitude",  "longitude", "altitude", "speed","direction","note"
    fra_tmp01 = pd.DataFrame(columns=["time",
                                      "vehicleNo", "vehicleName",
                                      "latitude",  "longitude",
                                      "altitude",  "speed",       "direction",
                                      "note"])
    for kind in arg_lstKind:
    #---- 한가지 종류의 탈것에 대한 데이터 생성 반복 ------------------------------
        
        #---- 한가기 종류 탈것의 댓수별 번호/이름 ---------------------------------
        lst_number  = np.arange(arg_lstSize[0]) + np.random.randint(1,9)*1000
        lst_vhcNo   = [str(num) + kind for num in lst_number]
        lst_vhcName = [kind + str(num) for num in lst_number]
    
        for oneVhcNo, oneVhcName in zip(lst_vhcNo, lst_vhcName):
        #---- 한대의 data 싱성 반복 ----------------------------------------------
            
            #---- time ----------------------------------------------------
            lst_dtTime = genOneVhcTimeLst(arg_lstDateRange[0],
                                          arg_lstDateRange[1],
                                          arg_lstSize[1])
#            print(f"lst_time\t: {len(lst_time)}\t{lst_time}")
        
            #---- latitude ----------------------------------------------------
            lst_latitude=[]
            
            delta = np.random.randint(1, 10)/100
            if np.random.randint(2) == 1:
                delta = -delta
            
            latitude = arg_lstPos[0]
            
            n=0
            while n < arg_lstSize[1]:
                latitude = latitude + delta
                latitude = round(latitude, 2)
                lst_latitude.append(latitude)
                n = n +1
#            print(f"lst_latitude\t: {len(lst_latitude)}\t{lst_latitude}")
            
            #---- longitude ----------------------------------------------------
            lst_longitude=[]
            
            delta = np.random.randint(1, 10)/100
            if np.random.randint(2) == 1:
                delta = -delta
                
            longitude = arg_lstPos[1]
            
            n=0
            while n < arg_lstSize[1]:
                longitude = longitude + delta
                longitude = round(longitude, 2)
                lst_longitude.append(longitude)
                n = n +1
#            print(f"lst_longitude\t: {len(lst_longitude)}\t{lst_longitude}")
            
            #---- altitude ----------------------------------------------------
            lst_altitude = []
            
            delta = np.random.randint(1, 10)
            if np.random.randint(2) == 1:
                delta = -delta
                
            altitude = 15000
            
            n=0
            while n < arg_lstSize[1]:
                altitude = altitude + delta
                altitude = round(altitude, 2)
                lst_altitude.append(altitude)
                n = n +1
#            print(f"lst_altitude\t: {len(lst_altitude)}\t{lst_altitude}")

            #---- speed ----------------------------------------------------
            lst_speed = []
        
            delta = np.random.randint(1, 10)
            if np.random.randint(2) == 1:
                delta = -delta
            
            speed = 250
            
            n=0
            while n < arg_lstSize[1]:
                speed = speed + delta
                speed = round(speed, 2)
                lst_speed.append(speed)
                n = n +1
#            print(f"lst_speed\t: {len(lst_speed)}\t{lst_speed}")

            #---- direction ----------------------------------------------------
            lst_dir = []
        
            delta = np.random.randint(1, 10)/10
            if np.random.randint(2) == 1:
                delta = -delta
            
            dir = 90
            
            n=0
            while n < arg_lstSize[1]:
                dir = dir + delta
                dir = round(dir, 2)
                lst_dir.append(dir)
                n = n +1
#            print(f"lst_dir\t\t: {len(lst_dir)}\t{lst_dir}")


    # "time","vehicleNo", "vehicleName","latitude",  "longitude", "altitude", "speed","direction","note"
        
            fra_part = pd.DataFrame({"time"         : lst_dtTime,
                                     "kind"         : [kind]         * arg_lstSize[1],
                                     "vehicleNo"    : [oneVhcNo]     * arg_lstSize[1],
                                     "vehicleName"  : [oneVhcName]   * arg_lstSize[1],
                                     "latitude"     : lst_latitude,
                                     "longitude"    : lst_longitude,
                                     "altitude"     : lst_altitude,
                                     "speed"        : lst_speed,
                                     "direction"    : lst_dir,
                                     "note"         : ["information"] * arg_lstSize[1]})
            
            fra_tmp01=pd.merge(fra_tmp01, fra_part, how="outer")
            
    fra_tmp01 = fra_tmp01.set_index("time")
    fra_tmp01 = fra_tmp01.sort_index()
    return(fra_tmp01)

fra_data = genVehicleDatas(["ADSB", "SHIP", "SQUAWK"], [100,20], ["2019-12-1", "2019-12-30"])

print(fra_data.head())
fra_data.to_csv("datasets/rundata.csv")
```
